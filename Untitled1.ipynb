```python

1
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Downloads\company-sales (1).csv")
total_profit = df['total_profit']
month_number = df['month_number']

plt.figure(figsize=(15, 6))
plt.plot(month_number, total_profit, linestyle='dotted', color='green', marker='s')
for i in range(len(df)):
    plt.text(df['month_number'][i], df['total_profit'][i], f'{df["total_profit"][i]}', ha='right')
plt.xticks(rotation=0)
plt.yticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True, alpha=0.3)
plt.title("Monthly Total Profit")
plt.show()

bathingsoap = df['bathingsoap']
facewash = df['facewash']

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.plot(month_number, bathingsoap, color='red', marker='o')
for i in range(len(df)):
    plt.text(df['month_number'][i], df['bathingsoap'][i], f'{df["bathingsoap"][i]}', ha='right')
plt.xticks(rotation=0)
plt.yticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Bathing Soap Sales')
plt.title("Bathing Soap Sales")
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(month_number, facewash, color='blue', marker='o')
for i in range(len(df)):
    plt.text(df['month_number'][i], df['facewash'][i], f'{df["facewash"][i]}', ha='right')
plt.xticks(rotation=0)
plt.yticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Face Wash Sales')
plt.title("Face Wash Sales")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

2:


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Downloads\company-sales (1).csv")
total_profit = df['total_profit']
month_number = df['month_number']

plt.figure(figsize=(15, 6))
plt.plot(month_number, total_profit, linestyle='dotted', color='green', marker='s', label='Total Profit')
for i in range(len(df)):
    plt.text(df['month_number'][i], df['total_profit'][i], f'{df["total_profit"][i]}', ha='right')
plt.xticks(rotation=0)
plt.yticks(rotation=45)
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True, alpha=0.3)
plt.legend(title='label', loc='upper right')
plt.show()

toothpaste = df['toothpaste']
plt.bar(month_number, toothpaste, color='skyblue', edgecolor='black')
plt.title('Toothpaste sales')
plt.xlabel('Month')
plt.ylabel('Toothpaste')
plt.grid(True, axis='y', alpha=0.3)
plt.show()

3


import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Downloads\company-sales (1).csv")
data = {
    'facecream': df['facecream'].sum(),
    'facewash': df['facewash'].sum(),
    'toothpaste': df['toothpaste'].sum(),
    'bathingsoap': df['bathingsoap'].sum(),
    'shampoo': df['shampoo'].sum(),
    'moisturizer': df['moisturizer'].sum()
}

max_sales = max(data.values())
max_product = [product for product, sales in data.items() if sales == max_sales][0]
total_profit = sum(data.values())

fig, ax = plt.subplots()
explode = [0.1 if sales == max_sales else 0 for sales in data.values()]
labels = [f'{product}\n{sales}' for product, sales in data.items()]
ax.pie(data.values(), explode=explode, labels=labels, startangle=60, autopct='%1.1f%%')
ax.set_title('Sales Data')
ax.axis('equal')
ax.legend(bbox_to_anchor=(0.125, 0.125))
plt.text(0.3, -1.5, f'Max Product: {max_product}\nTotal Profit: {total_profit}',
         horizontalalignment='center', verticalalignment='center')
plt.show()

facecream = df['facecream']
facewash = df['facewash']
month_number = df['month_number']
width = 0.4

plt.barh(month_number - 0.2, facecream, color='skyblue', label='Face Cream', height=width)
plt.barh(month_number + 0.2, facewash, color='lightgreen', label='Face Wash', height=width)
plt.xlabel('Sales')
plt.ylabel('Month')
plt.title('Sales for Face Cream and Face Wash')
plt.legend(loc='lower right')
plt.grid(True, axis='x', alpha=0.6)
plt.show()

 4


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

languages = ['Fict', 'Tech', 'Moti', 'Business', 'Nutri', 'Dev']
sale = [5.2, 19.6, 8.7, 8, 7.7, 3.7]
colors = plt.cm.inferno(np.linspace(0, 1, len(languages)))

bars = plt.barh(languages, sale, color=colors)
plt.xticks(range(0, 21, 2))
plt.yticks(languages)
for bar, lang in zip(bars, languages):
    bar.set_label(lang)
plt.xlabel('Sales')
plt.ylabel('Languages')
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3, axis='x')
plt.show()

categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
mean_men = [22, 30, 35, 35, 26]
mean_women = [25, 32, 30, 35, 29]
x = np.arange(len(categories))

plt.bar(x, mean_men, label='Men', color='maroon', edgecolor='black')
plt.bar(x, mean_women, bottom=mean_men, label='Women', color='gold', edgecolor='black')
for i in range(len(categories)):
    plt.text(x[i], mean_men[i] / 2, str(mean_men[i]), ha='center', color='white')
    plt.text(x[i], mean_men[i] + mean_women[i] / 2, str(mean_women[i]), ha='center', color='black')
plt.xlabel('Categories')
plt.ylabel('Mean')
plt.xticks(x, categories)
plt.grid(True, axis='y', alpha=0.6)
plt.legend(loc='upper right')
plt.show()

 5


import matplotlib.pyplot as plt
import pandas as pd

subjects = ['DSC', 'OOP', 'OPS', 'COA', 'MAT', 'Java']
pass_percentage = [40, 25.6, 8.8, 30, 7.7, 60.7]
explode = [0, 0, 0, 0, 0, 0.1]

plt.pie(pass_percentage, explode=explode, labels=subjects, startangle=140, autopct="%1.1f%%")
plt.legend(bbox_to_anchor=(1.2, 1.0))
plt.title("Pass Percentage by Subject")
plt.show()

df = pd.read_csv(r"C:\Users\Downloads\company-sales (1).csv")

plt.figure(figsize=(12, 6))
n, bins, patches = plt.hist(df['total_profit'], bins=10, edgecolor='black', color='green')
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title("Histogram of Total Profit")
plt.grid(True, axis='y', alpha=0.6)
plt.show()

for count, bin_start in zip(n, bins[:-1]):
    bin_end = bin_start + (bins[1] - bins[0])
    print(f'Range: [{bin_start:.2f} - {bin_end:.2f}] -> {int(count)} occurrences')

 6


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('planets.csv')

plt.figure(figsize=(15, 6))
scatter = sns.scatterplot(
    x='year',
    y='distance',
    hue='mass',
    style='number',
    size='mass',
    sizes=(20, 200),
    palette='inferno',
    data=df
)
scatter.set_title('Distance Covered Year-wise')
plt.legend(bbox_to_anchor=(1.2, 1.0))
plt.show()

plt.figure(figsize=(15, 6))
hist = sns.histplot(
    df,
    x='year',
    weights='orbital_period',
    bins=15,
    edgecolor='black',
    color='gold'
)
hist.set_title('Orbital Period Each Year')
plt.grid(True, axis='y', alpha=0.3)
plt.show()

```
